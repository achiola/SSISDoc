```python
import networkx as nx
twitter = nx.DiGraph()
twitter.add_node('you')
twitter.add_node('me')
twitter.add_edge('me','you') #order is important here.
```

```python
nx.draw(twitter, with_labels=True, font_weight='bold', node_size=1000)
```
