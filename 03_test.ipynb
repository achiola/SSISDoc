```python
# Import the required modules
import xmltodict
import pprint
  
# Open the file and read the contents
#with open('data/DIM_Clientes.dtsx', 'r', encoding='utf-8') as file:
with open('data/01_basico.dtsx', 'r', encoding='utf-8') as file:
    pkg_xml = file.read()
  
# Use xmltodict to parse and convert 
# the XML document
pkg_dict = xmltodict.parse(pkg_xml)
  
# Print the dictionary
# pprint.pprint(my_dict, indent=2)
```

```python
pkg_name = pkg_dict['DTS:Executable']['DTS:Property']
```

```python
pkg_properties = []
for p in pkg_dict['DTS:Executable']['DTS:Property']:
        lname = p['@DTS:Name']
        if(len(p) >= 2):
            lvalue = p['#text']
        else:
            lvalue = ''
        pkg_properties.append([lname, lvalue])
```

```python
conn_dic = {}
for p in pkg_dict['DTS:Executable']['DTS:ConnectionManager']:
    for i in p['DTS:Property']:
        if ( i['@DTS:Name'] == 'ObjectName' ):
            lname = i['#text']
        if ( i['@DTS:Name'] == 'DTSID' ):
            lid = i['#text']
    conn_dic[lid] = lname
        
```
