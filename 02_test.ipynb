```python
import xml.etree.ElementTree as ET
tree = ET.parse('data/DIM_Clientes.dtsx')
root = tree.getroot()
```

```python
root.tag
```

```python
for child in root:
    print('TAG:', child.tag)
    print('Att:', child.attrib)
    print('Value:', child.text )
    print('-----------')
```
